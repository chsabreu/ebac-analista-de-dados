## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'

# 1
print('# 1')
frases = cancao.split(',')
print(frases)
print(f'Tipo: {type(frases)}')

# 2
print('\n# 2')
palavra_Roda = cancao.index('Roda')
palavra_gigante = cancao.index('gigante')
print(f'Posição inicial da palavra Roda: {palavra_Roda}\nTipo: {type(palavra_Roda)}')
print(f'Posição inicial da palavra gigante: {palavra_gigante}\nTipo: {type(palavra_gigante)}')

# 3
print('\n# 3')
palavras_lower_case = cancao.islower()
print(f'Todos os caracteres são minúsculos?\nResposta: {palavras_lower_case}')
print(f'Tipo: {type(palavras_lower_case)}')
```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'

# taxa selic
frase_selic = noticia[0:noticia.index(';')]
print(f'Frase: {frase_selic} - {type(frase_selic)}')

taxa_selic = frase_selic.split(' ')[3]
print(f'Taxa selic: {taxa_selic} - {type(taxa_selic)}')

# ano
frase_ano = noticia[noticia.index(';')+2:len(noticia)]
print(f'\nFrase: {frase_ano} - {type(frase_ano)}')

valor_ano = frase_ano.split(' ')[5]
print(f'Valor do ano: {valor_ano} - {type(valor_ano)}')
```
